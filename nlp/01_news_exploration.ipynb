```python
import pandas as pd
import matplotlib.pyplot as plt
import string
```

```python
df = pd.read_json('gs://cytora-user-wout/News_Category_Dataset_v2.json', lines=True)
```

```python
df.head()
```

```python
df['category'].value_counts(ascending=True).plot('barh', figsize=(10,10));
```

```python
df['nr_of_words'] = df['headline'].apply(lambda x: len(x.split(' ')))
```

```python
df['nr_of_words'].value_counts().sort_index(ascending=False).plot('barh', figsize=(10,10));
```
